## 1. Improved Result: Correlation of Anxiety and Isolation

### Analysis Description:
To accurately estimate the relationship between psychological distress factors, we conducted a **Pearson correlation analysis** between 'anxiety' and 'isolation' scores, segmented by academic year (1st vs 4th year). 

To ensure the robustness of the correlation coefficient ($r$), we implemented **outlier removal using the Interquartile Range (IQR) method** on both variables. This process removed data points outside $1.5 \times IQR$, filtering out extreme noise and allowing for a more stable observation of how social-psychological links evolve over the university transition period.

## 2. Improved Result: Predicting Depression through PCA

### Analysis Description:
To address the poor generalizability (negative $R^2$) observed in the baseline model, we implemented a sophisticated analytical pipeline targeting 'depression' as the dependent variable:
* **Standardization**: We applied `StandardScaler` to 'academic_pressure' and 'average_sleep' to normalize their scales.
* **Dimensionality Reduction (PCA)**: We utilized **Principal Component Analysis (PCA)** to transform these two predictors into a single **'Stress Factor' (1st Principal Component)**. This reduced model complexity and mitigated multi-collinearity issues.
* **Robust Validation**: We evaluated the performance using **5-Fold Cross-Validation** to obtain an average $R^2$, providing a robust estimate of generalizability for our $N=87$ dataset.

## Summary and Conclusion

### 1. Psychological Insights
Our refined analysis confirms a strong correlation ($r \approx 0.75$) between anxiety and social isolation among **freshmen**. This relationship significantly weakens by the senior year, suggesting that social integration stressors are most acute during the initial transition to university life.

### 2. Strategic Recommendation
While academic pressure is a significant predictor of depression, its impact is best understood when combined with other stressors into a single latent factor (as shown by our PCA results). 

**Final Recommendation**: University counseling services should implement **early-intervention screenings for 1st-year students**, focusing specifically on the intersection of social isolation and academic workload to prevent the onset of depressive symptoms.
