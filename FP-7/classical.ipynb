```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import parse_data as ps

df = ps.get_data()
print("Data loaded for statistical analysis.")
```

```python
df_year1 = df[df['academic_year'] == '1st year']
df_year4 = df[df['academic_year'] == '4th year']

r1, p1 = stats.pearsonr(df_year1['anxiety'], df_year1['isolation'])
print(f"[1st Year] Correlation: {r1:.3f}, P-value: {p1:.4f}")

r4, p4 = stats.pearsonr(df_year4['anxiety'], df_year4['isolation'])
print(f"[4th Year] Correlation: {r4:.3f}, P-value: {p4:.4f}")
```

```python
plt.style.use('ggplot')
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.regplot(x='anxiety', y='isolation', data=df_year1, ax=axes[0], 
            line_kws={"color": "red"}, scatter_kws={'alpha': 0.6})
axes[0].set_title(f'Figure 1a: 1st Year (r={r1:.2f})')
axes[0].set_xlabel('Anxiety Score')
axes[0].set_ylabel('Isolation Score')

sns.regplot(x='anxiety', y='isolation', data=df_year4, ax=axes[1], 
            line_kws={"color": "blue", "linestyle": "--"}, scatter_kws={'alpha': 0.6})
axes[1].set_title(f'Figure 1b: 4th Year (r={r4:.2f})')
axes[1].set_xlabel('Anxiety Score')
axes[1].set_ylabel('Isolation Score')

plt.tight_layout()

plt.savefig('figure1_h2_comparison.png', dpi=300)
plt.show()

print("Figure 1 has been saved as 'figure1_h2_comparison.png'")
```
