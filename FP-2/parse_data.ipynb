# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df = pd.read_csv('MentalHealthSurvey.csv')

df.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- Depression(depression)
- Anxiety(anxiety)
- Isolation(isolation)

**IVs**:
- Academic Pressure(academic_pressure)
- Financial Concerns(financial_concerns)
- Social Relationships Score(social_relationships)


<br>
<br>

Let's extract the relevant columns:

```python
dv_cols = ['depression', 'anxiety', 'isolation']
iv_cols = ['academic_pressure', 'financial_concerns', 'social_relationships']
all_cols = dv_cols + iv_cols
df_parsed = df[all_cols].copy()
df = df_parsed
print(df.head())
df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = df.corr()
plt.figure(figsize=(10, 8)) 
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Variables')
plt.show()
```
