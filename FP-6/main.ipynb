**Figure 2: Relative Impact of Predictors on Depression Scores**
To identify predictors of depression, we implemented a Linear Regression model using an 80/20 train-test split for validation.
* **Validation Results**: The model achieved a Train R^2 of 0.162 and a Test R^2 of -0.048 on the novel test set.
* **Key Insight**: Despite the low predictive power, academic pressure has a significantly higher predictive weight compared to average sleep hours. This implies that subjective stress is a more critical indicator of mental health decline than physical rest duration.

## Summary and Conclusion

### 1. Psychological Decoupling
1st-year students exhibit a significant correlation between anxiety and isolation (r=0.731), which diminishes by the 4th year (r=0.424, P=0.222). This highlights the importance of early-stage social support for freshmen.

### 2. Predictors and Validation
Machine learning validation reveals that academic pressure is a primary driver of depression. Although the linear model shows limited generalizability (Test R^2 = -0.048), the identification of academic pressure as a dominant factor provides clear guidance for university counseling services.

### Final Recommendations
University support systems should prioritize targeted stress-management interventions for incoming 1st-year students to mitigate the impact of academic pressure on mental well-being.
