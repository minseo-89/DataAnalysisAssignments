### Figure 2

Use `data-fig2.npy` for this figure.

```python
b   = np.load( 'data-fig2.npy' )
plt.close()
plt.figure()
plt.plot(b[:,0], color='red', label='Red')
plt.plot(b[:,1], color='green', label='Green')
plt.plot(b[:,2], color='blue', label='Blue')
plt.xlabel("X value")
plt.ylabel("Y value")
plt.legend()
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
c   = np.load( 'data-fig3.npy' )
plt.close()
plt.figure()
plt.scatter(c[:, 0], c[:, 1], color='blue', alpha=0.35, s=200)
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
d   = np.load( 'data-fig4.npy' )
categories = ['A', 'B', 'C', 'D', 'E']
x_positions = np.arange(len(categories))
plt.close() 
plt.figure()
plt.bar(x_positions, d, color='0.7')
plt.xlabel("Category")
plt.ylabel("Value")
plt.xticks(x_positions, categories)
plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
e   = np.load( 'data-fig5.npy' )
plt.close()
plt.figure()
image = plt.imshow(e, cmap='RdBu', origin='lower',vmin=-1.0, vmax=1.0, extent=[0,40,0,120])
plt.colorbar(image)
x_ticks = np.arange(0,41,5)
plt.xticks(x_ticks)
y_ticks = np.arange(0,121,20)
plt.yticks(y_ticks)
plt.gca().set_aspect(0.333)
plt.show()
```
