# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('MentalHealthSurvey.csv')

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- depression

**IVs**:
- academic_pressure
- financial_concerns
- social_relationships


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['depression', 'academic_pressure', 'financial_concerns', 'social_relationships']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'depression': 'Depression',
    'academic_pressure': 'Pressure',
    'financial_concerns': 'Money',
    'social_relationships': 'Social'} )

df.describe()
```
